```python
import numpy as np
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
```

```python
DataDf = pd.read_excel('data.xlsx')
DataDf
```

```python
DataDf['Fecha']=DataDf['Fecha'].map(lambda x: datetime.strptime(x, "%d/%m/%Y"))
```

```python

fig, axes = plt.subplots(3, 1, sharey=True, figsize=(10,15)) # 3 rows, 1 column, shared y-axis

# Plot data on each subplot
axes[0].plot(DataDf[DataDf.columns[0]],DataDf[DataDf.columns[1]],color="blue")
#axes[0].xlabel(DataDf.columns[0])
axes[0].set_title(DataDf.columns[0]+ ' vs '+ DataDf.columns[1])
axes[0].grid(True)

axes[1].plot(DataDf[DataDf.columns[0]],DataDf[DataDf.columns[2]],color="green")
#axes[1].xlabel(DataDf.columns[0])
axes[1].set_title(DataDf.columns[0]+ ' vs '+ DataDf.columns[2])
axes[1].grid(True)

axes[2].plot(DataDf[DataDf.columns[0]],DataDf[DataDf.columns[3]],color="red")
#axes[2].xlabel(DataDf.columns[0])
axes[2].set_title(DataDf.columns[0]+ ' vs '+ DataDf.columns[3])
axes[2].grid(True)

plt.tight_layout() # Adjust layout to prevent overlapping titles/labels
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import savgol_filter

# Datos de ejemplo (reemplaza con tus propios datos)
x = DataDf[DataDf.columns[0]]
y = DataDf[DataDf.columns[1]]


# Parmetros del filtro Savitzky-Golay
window_length = len(DataDf)-1  # Longitud de la ventana (debe ser impar)
polyorder = 3      # Orden del polinomio

# Suavizar los datos
y_smooth = savgol_filter(y, window_length, polyorder)

# Graficar los datos originales y suavizados
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Datos originales', alpha=0.5)
plt.plot(x, y_smooth, label='Datos suavizados', linewidth=2)
plt.xlabel('Tiempo [Das]')
plt.ylabel('Ingreso bruto[COP]')
plt.xticks(rotation=45)
plt.title('Grfico Ingreso bruto')
plt.legend()
plt.grid(True)
plt.show()
```
